```r
#####################
### Octave kernel ###
#####################
# read input matrix #
#####################
W = dlmread("example_network_M_SD_010.txt")
```

```r
####################################
########## Octave kernel ###########
####################################
# compute some network descriptors #
####################################
rows=size(W,1) # number of rows (plants)
columns=size(W,2) # number of columns (animals)
interactions=nnz(W) # number of interactions
connectance=interactions/(rows*columns) # connectance
```

```r
#########################
#### Octave kernel ######
#########################
# compute NODF for rows #
#########################
G=zeros((rows*(rows-1))/2,3); # number of different pairs of rows
x=0;
for i=1: rows
    for j=i+1: rows
        suma=0;
        for k=1: columns
            suma=suma+(W(i,k)*W(j,k));
        end
        p=nnz(W(i,:)); # number of ones in the i row
        q=nnz(W(j,:)); # number of ones in the j row
        if p > q
            min_row=q;
            x=x+1;
            G(x,1)=i;
            G(x,2)=j;
            G(x,3)=suma/min_row;
        elseif p < q
            min_row=p;
            x=x+1;
            G(x,1)=i;
            G(x,2)=j;
            G(x,3)=suma/min_row;
        else # the contribution to NODF by a pair of rows with the same degree is zero
            x=x+1;
            G(x,1)=i;
            G(x,2)=j;
            G(x,3)=0;
        end
    end
end
G;
row_vector=sum(G,1); % sum of the elements of each column (size= 1 X 3)
result_rows=row_vector(1,3); # the third column indicates the contribution to NODF of all pairs of rows
x; # number of pairs of rows to compare
nodf_rows = result_rows/x
```

```r
############################
###### Octave kernel #######
############################
# compute NODF for columns #
############################
G=zeros((columns*(columns-1))/2,3); # number of different pairs of columns
y=0;
for i=1: columns
    for j=i+1: columns
        suma=0;
        for k=1: rows
            suma=suma+(W(k,i)*W(k,j));
        end
        p=nnz(W(:,i)); # number of ones in the i column
        q=nnz(W(:,j)); # number of ones in the j column
        if p > q
            min_column=q;
            y=y+1;
            G(y,1)=i;
            G(y,2)=j;
            G(y,3)=suma/min_column;
        elseif p < q
            min_column=p;
            y=y+1;
            G(y,1)=i;
            G(y,2)=j;
            G(y,3)=suma/min_column;
        else # the contribution to NODF by a pair of columns with the same degree is zero
            y=y+1;
            G(y,1)=i;
            G(y,2)=j;
            G(y,3)=0;
        end
    end
end
G;
col_vector=sum(G,1); # sum of the elements of each column (size= 1 X 3)
result_cols=col_vector(1,3); # the third column indicates the contribution to NODF of all pairs of columns
y; # number of pairs of columns to compare
nodf_cols = result_cols/y
```

```r
######################################
########## Octave kernel #############
######################################
# compute NODF for the entire matrix #
######################################
x;
nodf_rows;
y;
nodf_cols;
NODF = (result_rows+result_cols)/(x+y) # as in Almeida-Neto et al. (contribution to nodf of pairs with the same degree is not zero)
```
